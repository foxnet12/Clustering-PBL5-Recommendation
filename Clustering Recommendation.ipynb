##  Importing Tables from the Database and Merging into a Single Data Frame

```python
conn = sqlite3.connect("clustering.db")
```

```python
query = '''
SELECT *
FROM customers
JOIN products ON customers."Customers.id" = products."Products.id"
JOIN orders ON customers."Customers.id" = orders."Orders.customer_id"
'''
```

```python
merged_df = pd.read_sql_query(query, conn)
```

## Segmenting Customers using KMeans and Highlighting Characteristics

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
features = ['Customers.create_date', 'Customers.sales_rep', 'Orders.subtotal', 'Orders.total']
```

```python
X = merged_df[features].fillna(0)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
```

```python
merged_df['Cluster'] = kmeans.labels_
```

## Draw

```python
plt.scatter(X['Customers.create_date'], X['Orders.total'], c=kmeans.labels_)
plt.xlabel('Customers.create_date')
plt.ylabel('Orders.total')
plt.title('Customer Segmentation')
plt.show()
```

## Highlight characteristics of each cluster

```python
cluster_centers = kmeans.cluster_centers_
for i, center in enumerate(cluster_centers):
    print(f"Cluster {i+1} Center:")
    for j, feature in enumerate(features):
        print(f"{feature}: {center[j]}")
    print()
```
